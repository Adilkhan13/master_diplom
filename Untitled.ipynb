````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df = pd.read_excel('.\datasets\ipccc_industrial.xlsx')
````

````python
df.info()
````

````python
df['2.  Industrial Processes'] = df['2.  Industrial Processes'].replace('NO,NE,NA','0').astype(float)
````

````python

for tp in df['type'].unique():
    x = df[df['type'] == tp]['GREENHOUSE GAS SOURCE AND SINK CATEGORIES'].values
    y = df[df['type'] == tp]['2.  Industrial Processes'].values
    
    model = LinearRegression().fit(x.reshape(-1, 1), y)
    x_ = np.array([i for i in range(2019,2036)]).reshape(-1, 1)
    y_ = model.predict(x_)
    plt.figure(figsize=(15,5))
    plt.title(tp)
    plt.plot(x,y,label = tp)
    plt.plot(x_,y_,'-.')
    
    window = 5
    n = 30
    y = df[df['type'] == tp]['2.  Industrial Processes']
    roll_mean = y.rolling(window).mean()
    roll_std = y.rolling(window).std()

    #plt.plot(y[:n], label = tp, color = 'red')
    plt.plot(x,roll_mean[:n].to_list(),'-.', label = 'Скользящее среднее', color = 'blue')
    plt.plot(x,roll_std[:n].to_list(),'.', label = 'Скользящее стандартное отклонение', color = 'green')
    #plt.title("Скользящее среднее и скользящее стандартное отклонение для первых %d наблюдений"%(n))
    plt.legend(loc = 'best')
    plt.show(block = False)

    #plt.show()
````
